```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data=pd.read_csv('/content/mail_data.csv')
```

```python
raw_mail_data.head()
```

```python
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),'') #replacing null values with empty string
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
#Label Encoding ("spam"=0 and "ham"=1)
mail_data.loc[mail_data['Category']=="spam",'Category',]=0
mail_data.loc[mail_data['Category']=="ham",'Category',]=1
```

```python
mail_data.head()
```

```python
X=mail_data['Message']
Y=mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting data into Training data and Testing data
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.3,random_state=3)
#random_state meaning if you run the code multiple times with the same random_state, you will get exactly the same split every time.
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
#Feature Extraction: transform text data to numerical data fr LogisticRegression
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)

#TfidfVectorizer is a feature extraction tool from scikit-learn used in Natural Language Processing (NLP) to convert text data into numerical features that machine learning models
```

```python
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

```python
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
#Training the model
model=LogisticRegression()
```

```python
#Fitting training data with model
model.fit(X_train_features,Y_train)
```

```python
#Evaluating training model
prediction_on_train_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_train_data)
```

```python
print("Accuray on training data: ",accuracy_on_training_data)
```

```python
#prediction on test data
prediction_on_test_data=model.predict(X_test_features)
accuracy_on_testing_data=accuracy_score(Y_test,prediction_on_test_data)
```

```python
print("Accuray on testing data: ",accuracy_on_testing_data)
```

```python
#Building a predictive system
input_mail=["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]
input_data_features=feature_extraction.transform(input_mail)

#making prediction
prediction=model.predict(input_data_features)
if prediction[0]==1:
  print("Ham mail")
else:
  print("Spam mail")
```
